```python
"""
THE PROBLEME :
Reward calculation :

If they were in bed for more than 5h AND sleeping for more than 50% then, give Favorite_food.

If they have over 90 years old , give Favorite_food regardless.

Otherwise give Hate_food.

"""
```

```python
import pandas as pd
import numpy as np

def get_data(size = 10_000_000):
    # Create Fake Dataset
    df = pd.DataFrame()
    df['age'] = np.random.randint(1, 100, size)
    df['time_in_bed'] = np.random.randint(1, 9, size)
    df['pct_sleeping'] = np.random.rand(size)    
    df['Favorite_food'] = np.random.choice(['pizza','taco','ice-cream'], size)
    df['Hate_food'] = np.random.choice(['brocoli','candy corn','eggs'], size)
    return df
```

```python
df= get_data()
```

```python
mask_1 = ((df['pct_sleeping'] > 0.5) & (df['time_in_bed'] > 5)) | (df['age'] > 90)
mask_2 = (df['age'] < 20)
```

```python
df['reward'] = df['Hate_food']
# df.loc[((df['pct_sleeping'] > 0.5) & (df['time_in_bed'] > 5)) | (df['age'] > 90), 'reward'] = df['Favorite_food']
df.loc[mask_1, 'reward'] = df['Favorite_food']
df.loc[mask_2, 'reward'] = "yes teenage"
```
